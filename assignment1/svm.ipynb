#### SVM provement

**有用的矩阵求导公式**

1. 上下互为转置，得到单位阵

$$
\frac{\mathrm{d} \boldsymbol{x}^T}{\mathrm{~d} \boldsymbol{x}}=\boldsymbol{I} ,\quad \frac{\mathrm{d} \boldsymbol{x}}{\mathrm{d} \boldsymbol{x}^T}=\boldsymbol{I}
$$

2. 带有系数矩阵，上下变量互为转置，得到系数矩阵（否则是系数矩阵的转置）

$$
\begin{array}{ll}
\frac{\mathrm{d} \boldsymbol{x}^T \boldsymbol{A}}{\mathrm{d} \boldsymbol{x}}=\boldsymbol{A} & \frac{\mathrm{d} \boldsymbol{A} \boldsymbol{x}}{\mathrm{d} \boldsymbol{x}^T}=\boldsymbol{A} \\
\frac{\mathrm{d} \boldsymbol{A} \boldsymbol{x}}{\mathrm{d} \boldsymbol{x}}=\boldsymbol{A}^T & \frac{\mathrm{d} \boldsymbol{x} \boldsymbol{A}}{\mathrm{d} \boldsymbol{x}}=\boldsymbol{A}^T
\end{array}
$$

3. 变量转置乘上本身，求导得到两倍

$$
\frac{\mathrm{d} \boldsymbol{x}^T \boldsymbol{x}}{\mathrm{d} \boldsymbol{x}}=2 \boldsymbol{x}
$$

4. 类二项式矩阵求导

$$
\frac{\mathrm{d} \boldsymbol{x}^T \boldsymbol{A} \boldsymbol{x}}{\mathrm{d} \boldsymbol{x}}=\left(\boldsymbol{A}+\boldsymbol{A}^T\right) \boldsymbol{x}
$$

5. $$
   \frac{\partial \boldsymbol{u}^T \boldsymbol{X} \boldsymbol{v}}{\partial \boldsymbol{X}}=\boldsymbol{u} \boldsymbol{v}^T
   $$

6. $$
   \frac{\partial \boldsymbol{u}^T \boldsymbol{X}^T \boldsymbol{X} \boldsymbol{u}}{\partial \boldsymbol{X}}=2 \boldsymbol{X} \boldsymbol{u} \boldsymbol{u}^T
   $$



**对于svm公式的简单推导**

首先给出multi-class svm损失函数定义，对于每个数据样本，它的svm loss为：
$$
L_i=\sum_{j \neq y_i}\left[\max \left(0, \omega_j^T x_i-\omega_{y_i}^T x_i+\Delta\right)+\lambda \|W\|^2_2\right]
$$
这里的$\Delta$常设置为1.  然后我们求它对于参数w的梯度（求第i个样本对于第j个W参数的导数），采用分段求导的方式：
$$
\nabla_\omega L_{i, j}=\frac{\partial}{\partial \omega} L_{i, j}=\left\{\begin{array}{l}
\frac{\partial}{\partial \omega_j} L_i\left(j \neq y_i\right)= x_i\\
\frac{\partial}{\partial \omega_{y_i}} L_i\left(j=y_i\right)=-x_i
\end{array}\right.
$$


在这里给出所有样本累加后的总的loss以及总的梯度（做一个平均化）：
$$
L=\frac{1}{N}\sum_{i=1}^{N} \sum_{j \neq y_i}\left[\max \left(0, \omega_j^T x_i-\omega_{y_i}^T x_i+\Delta\right)\right]+\lambda  \|W\|^2_2 \\
$$
若这样定义的话，正则项求导会多出一个系数2，一般来说我们会令\lambda除以系数2来进行抵消。（当然不除也是没关系的，这关系到自定义超参数C）
$$
L=\frac{1}{N}\sum_{i=1}^{N} \sum_{j \neq y_i}\left[\max \left(0, \omega_j^T x_i-\omega_{y_i}^T x_i+\Delta\right)\right]+\frac{\lambda}{2}  \|W\|^2_2 \\
grad = \frac{1}{N} \sum_{i=1}^N \frac{\partial}{\partial \omega} L_{i, j} +  \lambda * W
$$
